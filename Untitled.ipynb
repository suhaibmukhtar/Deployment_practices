```python
import re
import datetime 
def validate(name,address,contact,password,cpassword,dob):
    pattern1="^0[0-9]{9}"
    if not name.strip():
        print("Name field should not be empty")
        return -1
    if not contact.strip():
        print("Contact field should not be empty")
        return -1
    if not password.strip():
        print("Password field should not be empty")
        return -1
    if not cpassword.strip():
        print("Confirm Password field should not be empty")
        return -1
    if not dob.strip():
        print("DOB field should not be empty")
        return -1
    if password!=cpassword:
        print("passwords are not matching!")
        return -1
    regex_contact=re.compile(pattern1)
    res1=regex_contact.search(contact)
    if not res1:
        print("Please enter the contact in Valid Format start with 0 then upto 10 digits!")
        return -1
    regex_password=re.compile("^[a-zA-Z]+[@$&][0-9]+")
    res2=regex_password.search(password)
    if not res2:
        print("Please enter the password in Valid Format [a-z]@[0-9]!")
        return -1
    current=datetime.datetime.now().year
    yr=re.findall("[0-9]{4}",dob)
    new_year=int(current)-int(yr[0])
    if new_year<21:
        return -1
    return 1

def main():
    while True:
        try:
            print("1.Signup\n2.Signin\n3.Quit")
            choice=int(input('Enter your choice'))
            if not choice:
                print("please enter the valid choice!")
            if choice==1:
                name=input('Enter the name:')
                address=input('Enter your address or skip press enter:')
                contact=input('Enter 10 digit mobile number:')
                password=input('Enter password:please mention special symbol:')
                cpassword=input('Enter confirm passowrd:')
                dob=input('Enter the date: # DD-MM-YYYY:')
                ##if validation 
                val=validate(name,address,contact,password,cpassword,dob)
                if val==1:
                    users={"name":name,'contact':contact,'password':password}
                    u1=User(name,address,contact,password,cpassword,dob)
                    print("Signup Successfull")
                    
                else:
                    print("Please enter details in valid format!")
            elif choice==2:
                if len(users)==0:
                    print('Signup first')
                else:
                    cont=input('Enter the username:(10 digit mobile number):')
                    pas=input('Enter the password:')
                    if u1.contact==cont and u1.password==pas:
                        print("Login Successfull")
                        ch=int(input("1.Press 1 for ordering\n2.Press 2 for Order Status\n3.Press 3 for previous menu\nEnter your choice:"))
                        if ch==1:
                            u1.dine()
                        elif ch==2:
                            print(u1.history)
                    else:
                        print("please enter valid details!")
        except ValueError:
            print("Invalid input. Please enter an integer.")
                
            
    
    
    
if __name__=="__main__":
    users={}
    main()
```
